```python
import numpy as np
import pandas as pd
from IPython.display import display
```

```python
def build_loc(category, year):
    year = str(year)
    dataloc =  f'../data/{category}/{category}_{year}.csv'
    metaloc =  f'../data/{category}/{category}_{year}_meta.csv'
    return dataloc, metaloc

def build_saveloc(category):
    saveloc = f'../data/clean/{category}_clean.csv'
    return saveloc
```

# Education Data 

```python
loc, metaloc = build_loc('education', 2010)
```

```python
data = pd.read_csv(loc, low_memory=False)
metadata = pd.read_csv(metaloc)
```

```python
data.head()
```

```python
metadata.head()
```

```python
tokeep_education = ['GEO_ID', 'NAME']
datacols = [s for s in metadata.iloc[14:27, 0].values if 'E' in s]
tokeep_education.extend(datacols)
tokeep_education
```

```python
data_small = data[tokeep_education]
```

```python
data_small.head()
```

```python
education_dfs = []
for yr in range(2010, 2020):
    loc, metaloc = build_loc('education', yr)
    data = pd.read_csv(loc, low_memory=False)
    data_small = data[tokeep_education]
    data_small = data_small.rename(columns=data_small.iloc[0]).drop(data_small.index[0])
    data_small = data_small.rename(columns=lambda s: s.split('!!')[-1].lower())
    data_small['year'] = np.repeat(yr, len(data_small))
    education_dfs.append(data_small)
```

```python
education_all = pd.concat(education_dfs, ignore_index=True)
education_all
```

```python
saveloc = build_saveloc('education')
# education_all.to_csv(saveloc, index=False)
```
